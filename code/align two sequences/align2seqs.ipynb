```python
def tonumber(letters):
    Dic={'A':1,'T':2,'G':3,'C':4}
    aa=[]
    for i in list(letters):
        aa.append(Dic[i])
    return(aa)
```

```python
def wrap0(x,b):
    x=x % (2*b)
    if x>b:
        x=2*b-x
    return(x)
def wrap(x,a,b):
    return(a+wrap0(x-a,b-a))
def showone(letters,path,S):
    row=[]
    x=0
    for i in range(len(path)):
        if path[i]==S:
            row.append('-')
        else:
            row.append(letters[x])
            x=x+1
    return(row)
def draw(letters1,letters2,path):
    print(showone(letters1,path,I))    
    print(showone(letters2,path,D))
```

```python
from numpy import *
import random
from __future__ import division 
def choose(pr):
    pr=pr/sum(pr)
    u=random.uniform(0, 1)
    total=0.0
    for i in range(len(pr)):
        total=total+pr[i]
        s=i+1
        if u< total:
            break
    return(s)
```

```python
import numpy
def pairhmm(d,e):
    x1 = numpy.array((1-2*d, d, d, 0, 1))
    x2 = numpy.array(((1-e)*(1-2*d), e + (1-e)*d, (1-e)*d, 0, 1))
    x3 = numpy.array(((1-e)*(1-2*d), (1-e)*d, e + (1-e)*d, 0, 1))
    x= numpy.array((x1, x2, x3))
    return(x)
Q= numpy.array([[-3,1,1,1],[1,-3,1,1],[1,1,-3,1],[1,1,1,-3]])/3
```

```python
def totalprob(FM):
    XX=FM.shape[1]-1
    YY=FM.shape[2]-1
    return(FM[0,XX,YY] + FM[1,XX,YY] + FM[2,XX,YY])
```

```python
def backsample(FM,T):
    X=FM.shape[1]
    Y=FM.shape[2]
    x=X
    y=Y
    s=E
    states=[]
    while x>1 or y>1:
        s=choose(FM[:,x-1,y-1]*T[:,s-1])
        states.append(s)
        if s == 1:
            x=x-1
            y=y-1
        elif s == 2:
            x=x-1
        elif s == 3:
            y=y-1
        else:
            print("hello world")
    return(states)
```

```python
import math
t=0.25
d=0.05
e=0.5
P=power(math.e,Q*t)-(ones((4,4))-eye(4))
pi=[0.25,0.25,0.25,0.25]
```

```python
def forwardmatrix(seq1,seq2,T,P,pi):
    X=1+len(seq1)
    Y=1+len(seq2)
    FM=numpy.zeros((3,X,Y))
    FM[0,0,0] = 1
    for x in range(1,X+1):
        for y in range(1,Y+1):
            lx=seq1[x-2]
            ly=seq2[y-2]
            if x>1 and y>1:
                x2=x-2
                y2=y-2
                FM[0,x-1,y-1]=FM[0,x2,y2]*T[0,0]+FM[1,x2,y2]*T[1,0]+FM[2,x2,y2]*T[2,0]
                FM[0,x-1,y-1]=FM[0,x-1,y-1]*pi[lx]*P[lx,ly]
            if x>1:
                x2=x-2
                y2=y-1
                FM[1,x-1,y-1]=FM[0,x2,y2]*T[0,1]+FM[1,x2,y2]*T[1,1]+FM[2,x2,y2]*T[2,1]
                FM[1,x-1,y-1]=FM[1,x-1,y-1]*pi[lx]
            if y>1:
                x2=x-1
                y2=y-2
                FM[2,x-1,y-1]=FM[0,x2,y2]*T[0,2]+FM[1,x2,y2]*T[1,2]+FM[2,x2,y2]*T[2,2]
                FM[2,x-1,y-1]=FM[2,x-1,y-1]*pi[ly]
            else:
                pass
    return(FM)
```
